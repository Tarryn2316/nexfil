```python
import sys
import subprocess

process = subprocess.Popen([sys.executable, "nexfil.py", "-u", "niko.karhu"], stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

while True:
    output = process.stdout.readline()
    if output == '' and process.poll() is not None:
        break
    if output:
        print(output.strip())
```
